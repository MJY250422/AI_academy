```python
file_path = '/content/drive/MyDrive/Dataset/광주광역시_대기중 중금속 조사_20250411.csv'
df = pd.read_csv(file_path, encoding= 'euc-kr')
```

```python
df['지점_통합'] = df['지점'].str.extract(r'([가-힣]+)')
df = df.dropna()
df.head(20)
```

```python
heavy_metal_columns = [
    '납(Pb)', '카드뮴(Cd)', '크롬(Cr)', '구리(Cu)', '망간(Mn)', '철(Fe)',
    '니켈(Ni)', '비소(As)', '베릴륨(Be)', '알루미늄(Al)', '칼슘(Ca)', '마그네슘(Mg)'
]

print(df[heavy_metal_columns].mean())
print('-'*50)
print(df[heavy_metal_columns].median())
print('-'*50)
print(df[heavy_metal_columns].std())
```

```python
numeric_columns = df.select_dtypes(include=np.number).columns.tolist()

for col_name in numeric_columns: # 'col_name' 변수를 사용하여 컬럼 이름에 접근
        mean_value = df[col_name].mean() # 대괄호 []를 사용하여 컬럼에 접근
        print(f'{col_name}의 평균: {mean_value:.4f}') # f-string에서 변수 이름을 올바르게 사용
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas as pd
import numpy as np

# --- 1. df와 '총 중금속' 컬럼이 이미 준비되었다고 가정합니다. ---
#    (사용자 환경에서 파일 로드 및 전처리, '총 중금속' 컬럼 생성이 완료된 상태)

# 예를 들어, 만약 이 셀만 따로 실행한다면 아래 코드가 필요합니다.
# 하지만 사용자 환경에서는 이미 df와 '총 중금속' 컬럼이 있으므로 주석 처리합니다.
# file_path = '/content/drive/MyDrive/Dataset/광주광역시_대기중 중금속 조사_20250411.csv'
# df = pd.read_csv(file_path, encoding='euc-kr')
# df['지점_통합'] = df['지점'].str.extract(r'([가-힣]+)')
# df = df.dropna()
# heavy_metal_columns = [
#     '납(Pb)', '카드뮴(Cd)', '크롬(Cr)', '구리(Cu)', '망간(Mn)', '철(Fe)',
#     '니켈(Ni)', '비소(As)', '베릴륨(Be)', '알루미늄(Al)', '칼슘(Ca)', '마그네슘(Mg)'
# ]
# existing_heavy_metal_columns = [col for col in heavy_metal_columns if col in df.columns]
# if existing_heavy_metal_columns:
#     df['총 중금속'] = df[existing_heavy_metal_columns].sum(axis=1)
# else:
#     print("오류: '총 중금속'을 계산할 중금속 컬럼을 찾을 수 없습니다. ANOVA 진행 불가.")
#     exit()

print("\n--- ANOVA 분석 준비: '지점_통합' 그룹 정보 ---")
group_counts = df['지점_통합'].value_counts()
print(group_counts)

# ----------------------------------------------------
# 2. ANOVA 분석 코드
# ----------------------------------------------------

# ANOVA를 위한 최소 그룹 수 조건 확인 (최소 3개 그룹)
if df['지점_통합'].nunique() < 3:
    print(f"\n[오류]: '지점_통합' 컬럼의 고유 그룹 수가 {df['지점_통합'].nunique()}개로, ANOVA를 수행하기에 부족합니다 (최소 3개 그룹 필요).")
    print("그룹 수를 늘리거나, 그룹을 재정의해야 ANOVA 분석이 가능합니다.")
elif (group_counts < 2).any(): # 각 그룹에 최소 2개의 데이터가 있어야 함
    problematic_groups = group_counts[group_counts < 2].index.tolist()
    print(f"\n[오류]: 일부 '지점_통합' 그룹에 데이터가 1개뿐입니다 (예: {problematic_groups}).")
    print("ANOVA는 각 그룹 내 분산을 계산하므로, 최소 2개 이상의 데이터가 필요합니다. 해당 그룹을 제거하거나, 데이터를 추가해야 합니다.")
else:
    # OLS (Ordinary Least Squares) 모델 정의 및 학습
    # 종속 변수: '총 중금속', 독립 변수: '지점_통합'
    # 'C()'는 statsmodels에 '지점_통합'이 범주형 변수임을 알려줍니다.
    try:
        # 컬럼 이름에 공백이나 특수문자가 있을 경우, Q("")를 사용하여 감싸는 것이 안전합니다.
        model = ols('Q("총 중금속") ~ C(지점_통합)', data=df).fit()

        # ANOVA 테이블 생성 및 출력
        # typ=2는 Type II Sum of Squares를 의미하며, 불균형 데이터셋에 일반적으로 권장됩니다.
        anova_table = sm.stats.anova_lm(model, typ=2)

        print("\n--- ANOVA 분석 결과 ('총 중금속' vs '지점_통합') ---")
        print(anova_table)

        # ----------------------------------------------------
        # 3. 결과 해석
        # ----------------------------------------------------
        p_value = anova_table['PR(>F)']['C(지점_통합)']
        alpha = 0.05 # 유의수준 (일반적으로 0.05 사용)

        print(f"\n--- ANOVA 결과 해석 (유의수준: {alpha}) ---")
        if p_value < alpha:
            print(f"P-value ({p_value:.4f})가 유의수준 ({alpha})보다 작으므로, 귀무가설을 기각합니다.")
            print("결론: '지점_통합'에 따라 '총 중금속' 농도의 평균에 통계적으로 **유의미한 차이가 있다고 볼 수 있습니다.**")
            print("이는 적어도 하나의 지점 그룹에서 총 중금속 농도 평균이 다른 그룹과 다르다는 것을 의미합니다.")
            print("어떤 특정 그룹 간에 차이가 있는지 자세히 파악하려면 **사후 분석(Post-hoc test)**이 필요합니다. (예: Tukey HSD, Bonferroni 등)")
        else:
            print(f"P-value ({p_value:.4f})가 유의수준 ({alpha})보다 크거나 같으므로, 귀무가설을 기각할 수 없습니다.")
            print("결론: '지점_통합'에 따라 '총 중금속' 농도의 평균에 통계적으로 **유의미한 차이가 있다고 보기 어렵습니다.**")
            print("즉, 각 지점 그룹 간 총 중금속 농도 평균이 통계적으로 동일하다고 가정할 수 있습니다.")

    except Exception as e:
        print(f"\nANOVA 분석 중 오류가 발생했습니다: {e}")
        print("데이터프레임의 컬럼 이름('지점_통합', '총 중금속')이나 데이터 유형을 다시 확인해주세요.")
        print("특히, '총 중금속' 컬럼이 숫자형인지, '지점_통합' 컬럼이 범주형으로 인식되는지 확인해주세요.")
```
